Marine microalgae is increasingly recognized for its potential to meet societal needs at the food-energy-water nexus. Initially developed as a means for producing cheaper home-grown fuels to replace liquid transport fuels and thus promote energy security, marine microalgae have come to prove their worth far beyond a fuel replacement. Nonetheless, research has demonstrated the incredible bioproductivity of marine algae. In biofuel studies, it is found that microalgae has a fuel productivity of 0.50 bbl/ha/day. This means that microalgae require a land area of 1.92 million km2 to meet global liquid fuel demands in 2016, or land area that is slightly less than 3 times the size of Texas (Greene et al. 2016). Unlike traditional biofuels that compete with land used for traditional agriculture, microalgae production does not require arable land. The same land used to meet global liquid fuel demands would generate 2.40 Gt of protein co-product 10 times the annual global soy protein production (United Nations Food and Agriculture Organization, 2016). However, this is only a fraction of its potential; productivity estimates generally l assume that the microalgae strain cultivated provides a mixed equal output of biopetroleum production and food production at 1.25 million kg/km2 per year (Greene et al. 2016). Should production shift completely towards protein maximisation, the same amount of land could produce more than the aforementioned figure. 

Currently, the protein co-product is rendered into fish meal replacements for aquaculture (Henchion et al. 2017). Its benefit in aquaculture and terrestrial farming has been shown across many species, including sea bream, salmon, shrimp and poultry (Craigie 2010; O’Sullivan et al. 2010; Rajauria 2015; Heuzé et al. 2016; Makkar et al. 2016). Feeding trials at Cornell University found that chickens that consumed algae-supplemented feeds produced eggs that had three times the omega-3 fatty acids (Lum et al. 2013). Unfortunately, awareness of the health benefits of algae as a form of human nutrition “lags far behind that on which diets have been formulated for commercially important species and agriculture” (Wells et al. 2016). Thus, similar assimilation of microalgae into the human diet is still nascent and holds much potential for expansion.

CONCLUSION

Despite its potential for filling nutritional gaps, commodity-scale production of marine microalgae for food has yet to materialize. Technical challenges stand in the way of large scale microalgae cultivation including significant nitrogen and phosphorus inputs as well as CO2 for economic growth rates (Walsh 2017). However, microalgae presents a number of benefits, including high lipid content, that is worth pursuing. Additionally, marine microalgae, unlike current pharmaceutical freshwater algae products, could be cultivated with no freshwater requirement. Thus, there is potential to expand research of marine microalgae with the intention of scaling it up to a commodity (Wells et al., 2017). Additionally, research is needed to evaluate other health benefits of marine algae such as fucoidans, ulvans, alginate and antioxidants.

In particular, taste tests to determine consumer willingness to adopt marine microalgae protein in their consumption habits are needed. Product development of marine microalgae is vital in penetrating the commodity market. Currently, algae is found in food in powdered forms that are added to energy bars, baby formula, etc. (Wells et al. 2017). This could be useful as a powdered supplement to the vegetal grains commonly consumed as staples, which have been proven to facilitate greater absorption of iron from these grains. There could be potential in exploring more voluminous representations of marine algae to constitute a fuller meal. Industry could, for example, take the concept of pellets in aquaculture to aggregating microalgal biomass into more substantial food particles for consumption e.g. burger patties, or rice pellets. The first step will be, however, to see how microalage affects the taste and perception of commonly eaten foods.
